```python
import json
def load_json(path):
    """读取json文件"""
    with open(path, 'r', encoding='utf-8') as fp:
        data_list = json.load(fp)
    return data_list
ss = load_json('./sentences.json')
ts = load_json('./translated_sentences.json')
```

{"context": "Instruction: 将这句话改写成鲁迅风格的语言\nInput: 虽然打高尔夫、脱钩和参加豪门社交都是炫耀身份的行为，但如此疯狂追求名利地步就算被曝出丑闻，也不为过。\nAnswer: ", "target": "“打高尔夫”“脱钩”，“豪门社交”的熏灼之状，竟至于斯，则虽报以丑闻，亦不为过。"}

```python
import random

results = []

instructions = ['请把接下来的话语改写成鲁迅的风格','请用鲁迅的口吻改写下面的句子','请将以下文字转化为鲁迅的笔调','请模仿鲁迅的文风改写以下句子','请用鲁迅的语言风格重新表达以下内容','请将以下话语改写为鲁迅式的表达','请用鲁迅的表达方式重新撰写以下内容','请将以下句子转化为鲁迅的文学风格','请用鲁迅的语调改写以下文字','请将以下文字改写为鲁迅的经典风格']

for (org, trans) in zip(ss,ts):
    instruction = random.choice(instructions)    
    result = {"context": f"Instruction: {instruction}\nInput: {trans} \nAnswer: ",
              "target": f"{org}"
              }
    results.append(result)
```

```python
results
```

```python
with open("luxun_dataset.jsonl", 'w', encoding='utf-8') as f:
    for item in results:
        # 将每个字典转换为 JSON 字符串并写入文件
        json_line = json.dumps(item, ensure_ascii=False)
        f.write(json_line + '\n')
```
